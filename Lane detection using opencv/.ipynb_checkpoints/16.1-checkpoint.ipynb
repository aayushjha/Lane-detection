```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pylab import rcParams
rcParams["figure.figsize"]=(20,20)
```

```python
test=cv.imread("solidWhiteRight.jpg")
```

```python
test1=cv.cvtColor(test,cv.COLOR_BGR2RGB)
plt.imshow(test1)
```

```python
test2=cv.cvtColor(test1,cv.COLOR_RGB2GRAY)
plt.imshow(test2,cmap="gray")
```

```python
#gaussian blur
test_b=cv.GaussianBlur(test2,(5,5),0)
plt.imshow(test_b,cmap="gray")
```

```python
#canny edge detection
test_c=cv.Canny(test_b,120,225)
plt.imshow(test_c,cmap="gray")
```

```python
#generate a blackmask
mask=np.zeros_like(test_c)
p=np.array([(150,540),(400,330),(600,330),(900,540)])
reg=cv.fillPoly(mask,[p],255)
```

```python
plt.imshow(mask,cmap="gray")
```

```python
reg_b=cv.bitwise_and(test_c,reg)
```

```python
plt.imshow(reg_b,cmap="gray")
```

```python
lines=cv.HoughLinesP(
    reg_b,
    rho=6,
    theta=np.pi / 1800,
    threshold=20,
    minLineLength=10,
    lines=np.array([]),
    maxLineGap=5
)
```

```python
lines.shape
```

```python
for line in lines.reshape(-1,4):
    x1,y1,x2,y2=line
    cv.line(test,(x1,y1),(x2,y2),(255,255,0),10)
plt.imshow(test)
```

```python
cam=cv.VideoCapture("solidWhiteRight.mp4")
while True:
    try:
        s,frame=cam.read()
        test1=cv.cvtColor(frame,cv.COLOR_BGR2RGB)
        test2=cv.cvtColor(test1,cv.COLOR_RGB2GRAY)
        test_b=cv.GaussianBlur(test2,(5,5),0)
        test_c=cv.Canny(test_b,120,225)
        mask=np.zeros_like(test_c)
        p=np.array([(150,540),(400,330),(600,330),(900,540)])
        reg=cv.fillPoly(mask,[p],255)
        reg_b=cv.bitwise_and(test_c,reg)
        lines=cv.HoughLinesP(
            reg_b,
            rho=6,
            theta=np.pi / 1800,
            threshold=20,
            minLineLength=10,
            lines=np.array([]),
            maxLineGap=5
        )
        for line in lines.reshape(-1,4):
            x1,y1,x2,y2=line
            cv.line(frame,(x1,y1),(x2,y2),(0,200,200),10)
        cv.imshow("1",frame)
    except:
        pass
    
    key=cv.waitKey(1)
    if key==ord("q"):
        cv.destroyAllWindows()
        break
cam.release()
```
